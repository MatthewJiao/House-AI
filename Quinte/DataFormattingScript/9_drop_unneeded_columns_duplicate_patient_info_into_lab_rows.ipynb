```python
import pandas as pd
```

```python


df1 = pd.read_csv(
     '../9.csv',
)
```

```python
# 'Unnamed: 0', 'Unnamed: 0.1', 'Age_Code',
df1.drop(['Unnamed: 0', 'Unnamed: 0.1', 'Age_Code','Diag_Code'], axis = 1, inplace = True)
```

```python
df1.head(50)
```

```python
import math
ageDict = {}
genderDict = {}
weightDict = {}
diagnosisDict = {}
```

```python
print(diagnosisDict.get('H1012043'))
```

```python
df1.at[1, 'Age_Number'] = 81
```

```python
rows = df1.shape[0]
```

```python
# make dict for age // works
i = 0
while i < rows:
    if (df1.iloc[i][0] not in ageDict) and (math.isnan(df1.iloc[i][8]) == False):
        ageDict[df1.iloc[i][0]] = df1.iloc[i][8]
    i = i + 1
```

```python
# make dict for gender // works
i = 0
while i < rows:
    if (df1.iloc[i][0] not in genderDict) and (isinstance(df1.iloc[i][9], str) == True):
        genderDict[df1.iloc[i][0]] = df1.iloc[i][9]
    i = i + 1
```

```python
# make dict for weight // works
i = 0
while i < rows:
    if (df1.iloc[i][0] not in weightDict) and (math.isnan(df1.iloc[i][10]) == False):
        weightDict[df1.iloc[i][0]] = df1.iloc[i][10]
    i = i + 1
```

```python
# make dict for diagnosis // works
i = 0
while i < rows:
    if (df1.iloc[i][0] not in diagnosisDict):
        if isinstance(df1.iloc[i][11], str) == True:
            diagnosisDict[df1.iloc[i][0]] = df1.iloc[i][11]
    i = i + 1
```

```python
# fill in blanks
i = 0
while i < rows:
    # age
    if (math.isnan(df1.iloc[i][8]) == True) and (df1.iloc[i][0] in ageDict):
        df1.at[i, 'Age_Number'] = ageDict.get(df1.iloc[i][0])
    
    # gender
    if (isinstance(df1.iloc[i][9], str) == False) and (df1.iloc[i][0] in genderDict):
        df1.at[i, 'Gender'] = genderDict.get(df1.iloc[i][0])
    
    # weight
    if (math.isnan(df1.iloc[i][10]) == True) and (df1.iloc[i][0] in weightDict):
        df1.at[i, 'Weight'] = weightDict.get(df1.iloc[i][0])
    
    # diagnosis
    if (isinstance(df1.iloc[i][11], str) == False):
        if df1.iloc[i][0] in diagnosisDict:
            df1.at[i, 'Diag_desc'] = diagnosisDict.get(df1.iloc[i][0])
        else:
            df1.at[i, 'Diag_desc'] = 'Non sepsis patient'

    
    i = i + 1
```
